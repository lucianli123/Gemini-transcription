```python
for f in files:
    #id=f.split("\\")[-1].split(".")[0]
    id=f.split("/")[-2]+f.split("/")[-1]
    out_path="lines/"+id+".jpg"
    #print(out_path)
    if os.path.isfile(out_path):
        continue
    img = Image.open(f).convert('L')
    

    # Open the image

    # Create an ImageEnhance.Contrast object and specify the enhancement factor
    enhancer = ImageEnhance.Contrast(img)
    factor = 2 # Increase the contrast by 100%

    # Apply the enhancement
    img = enhancer.enhance(factor)
    img.save(out_path, quality=10) 
```
